```python
#import packages
import pandas as pd
```

```python
#read data from csv (URL)
df1 = pd.read_csv('https://dev.hsl.fi/citybikes/od-trips-2021/2021-05.csv')
df2 = pd.read_csv('https://dev.hsl.fi/citybikes/od-trips-2021/2021-06.csv')
df3 = pd.read_csv('https://dev.hsl.fi/citybikes/od-trips-2021/2021-07.csv')

#concat all dataframes together 
frames = [df1, df2, df3]
result = pd.concat(frames, ignore_index = True)

#print 
print(df1.shape)
print(df2.shape)
print(df3.shape)
display(result)
```

```python
#drop unnecissary info
drop = ['Departure station name', 'Return station name']
result = result.drop(drop, axis=1)

#print
result
```

```python
#change names of columns to fit our database names
result.columns = ['start_time', 'end_time', 'start_locatin_id', 'end_location_id', 'distance', 'duration']

#print
result
```

```python
# set missing values (NaN) as 0
result = result.fillna(0)
```

```python
#drop rows with distance under 10meters and duration under 10seconds
result = result.drop(result[(result['distance'] <= 10) | (result['duration'] <= 10)].index).reset_index(drop=True)

#print
result
```

```python
#export dataframe as csv 
csv = result.to_csv('citybike_ready_for_DB.csv', index=False)
```
